## Multiplication

```python
# Matrix multiply Vector

A = torch.arange(20, dtype = torch.float32).reshape(5,4)
x = torch.arange(4, dtype = torch.float32)

print(A, A.shape)
print(x, x.shape)
print(torch.mv(A, x), torch.mv(A, x).shape)
```

```python
# Matrix multiply Matrix

print(A, A.shape)

B = torch.ones(4, 3)
print(B, B.shape)

print(torch.mm(A, B), torch.mm(A, B).shape)  # torch.mm is strictly limited to 2-dim matrix calculation, can't be used on tensors.  
```

## Frobenius Norm

```python
# Frobenius Norm is the most commonly used norm of Matrices

M = torch.ones(4, 9)    
print(M)
print(torch.norm(M))  # sqrt(sum(square of each element))
```

# Tensor Calculation

Tensor can be regarded as the generalization of matrices.        
Matrices ususally have 2 dimision (row and column), whereas tensor can have any number of dimisions.     

## Tensor with Scalar

```python
# Add & Multiply scalar to Tensor

a = 2 
X = torch.arange(24).reshape(2,3,4) 

a + X, a * X, (a * X).shape   
```

```python
X = torch.arange(24).reshape(2, 3, 4)
X
```

## Tensor with Tensor

```python
# Add Tensors

A = torch.arange(20, dtype=torch.float32).reshape(5, 4)
B = A.clone()  # creates a new object, B is not a reference of A 

A, A + B
```

```python
# Multiply Tensors  

C = torch.arange(24, dtype=torch.float32).reshape(2, 3, 4)  # batch of 2 matrices, matrices shape = 3x4
D = torch.ones(2, 4, 5)                                     # batch of 2 matrices, matrices shape = 4x5
# batch number have to be same (2), matrices shape must satisfy matrix multiplication (3x4 * 4x5 = 3x5). Result will be 2x3x5

print(C, C.shape)
print(D, D.shape)
print(C @ D, (torch.matmul(C, D).shape))   # '@' is the magic command for torch.matmul()
```

```python
# Hadamard Product

print(A)
print(B)

print(A * B)    # A and B have same shape, calculate element-wise product
```

## Statistics

```python
# Sum of all elements

A = torch.arange(20, dtype=torch.float32).reshape(5, 4)
A, A.sum()
```

```python
# Sum by Dimension

X = torch.arange(24).reshape(2, 3, 4)
X, X.shape
```

```python
print(X.sum(axis=0))  # axis=0 --> eliminate the first dimension
print(X.sum(axis=0).shape)

print(X.sum(axis=1))  # axis=1 --> eliminate the second dimension
print(X.sum(axis=1).shape)

print(X.sum(axis=2))  # axis=2 --> eliminate the third dimension
print(X.sum(axis=2).shape)
```

```python
# Sum by multiple dimensions: 

print(X)
print(X.shape)

print(X.sum(axis=[0, 1]))  # axis = 0 & 1 --> eliminate the first & second dimension
print(X.sum(axis=[0, 1]).shape)

print(X.sum(axis=[1, 2]))  # axis = 1 & 2 --> eliminate the second & third dimension
print(X.sum(axis=[1, 2]).shape)

print(X.sum(axis=[0, 2]))  # axis = 0 & 2 --> eliminate the first & third dimension
print(X.sum(axis=[0, 2]).shape)
```

```python
# Calculate Statistics of Tensor

print(A.mean())
print(A.sum() / A.numel())
```

```python
# Calculate Statistics by Axis

print(A, A.shape)
print(A.mean(axis=0))     # calculate mean along axis=0, eliminate axis=0 (first dimension)
print(A.sum(axis=0)/A.shape[0])   # same effect as above line
print(A.mean(axis=1))     # calcualte mean along axis=1, eliminate axis=1 (second dimension)
```

```python
# Cumulative Sum

print(A)
print(A.cumsum(axis=0))
```

## Broadcast

```python
print(A)
print(A.shape)
```

```python
sum_A = A.sum(axis=1, keepdims=True)  # sum up each row (axis=1), keep the rows=5 and column=1
print(sum_A)                          # has same number of dimensions with original tensor A
print(sum_A.shape)                    # 5x1 can be used for broadcast to the original tensor 5x4
```

```python
A / sum_A   # divide each element by it's row's sum
```

```python
# False Example: 

print(A.sum(axis=1))        # only sum up by axis=1, # dimensions reduce by 1
print(A.sum(axis=1).shape)  # lose the column dimension, can't be used for broadcast as # dimensions changed

A / A.sum(axis=1)
```
